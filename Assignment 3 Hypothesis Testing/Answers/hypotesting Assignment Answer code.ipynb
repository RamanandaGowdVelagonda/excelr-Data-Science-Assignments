```python
import numpy as np
import pandas as pd
#from scipy.stats import chi2_contingen
d1 = {
    'Region': ['North', 'South', 'East', 'West'],
    'Male': [50, 142, 131, 70],
    'Female': [55, 351, 480, 350]
}
df2= pd.DataFrame(d1)
df2
```

```python
df2['Male'].mean()
df2['Female'].mean()
from scipy.stats import ttest_ind as twosampletest
zcal,pvalue = twosampletest(df2["Male"],df2["Female"])                             
print("zcalculated value: ",zcal.round(3))
print("pvalue :",pvalue.round(3))
```

```python
alpha = 0.05
if pvalue < alpha:
    print("Ho is rejected and H1 is accepted")
else:
    print("H1 is rejected and Ho is accepted")
```

```python
import pandas as pd
from scipy import stats

# Load the data
df3 = pd.read_csv("F:\Excler class\Assignment\LabTAT.csv")
df3

```

```python
# Perform ANOVA using iloc
f_statistic, p_value = stats.f_oneway(df3.iloc[:, 1], df3.iloc[:, 2], df3.iloc[:, 3])

# Set the alpha level
alpha = 0.05
# Compare the p-value to the alpha level
if p_value < alpha:
 print("Ho is rejected and H1 is accepted")
else:
    print("H1 is rejected and H0 is accepted")
print("F-statistic:", f_statistic)
print("p-value:", p_value)

```

```python
df4 = pd.read_csv("F:\Excler class\Assignment\Costomer+OrderForm.csv")
df4
```

```python
# Create a contingency table
data = [[15, 10, 5],
        [20, 15, 10],
        [25, 20, 15]]

# Create a DataFrame
df4 = pd.DataFrame(data,
                  index=["Center 1", "Center 2", "Center 3"],
                  columns=["Defective", "Non-defective", "Total"])
```

```python
from scipy.stats import chi2_contingency
# Perform a chi-square test
chi2, p, dof, expected = chi2_contingency(df4)

# Set the alpha level
alpha = 0.05

# Compare the p-value to the alpha level
if p < alpha:
    print("There is a significant difference in the proportion of defective order forms among the different centers.")
else:
    print("There is no significant difference in the proportion of defective order forms among the different centers.")
```
