# K-Means clustering

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("G:\my Assignemt\Assignment 7 Clustering\crime_data.csv")
data
```

```python
# normalizing
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data_df = scaler.fit_transform(data.iloc[:,1:])
scaled_data_df
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_data_df)
    wcss.append(kmeans.inertia_)
    if i>5:
        break
wcss      
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_data_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# Build Cluster Algorithm

clusters_new = KMeans(n_clusters=4,random_state=16)
clusters_new.fit(scaled_data_df)
```

```python
clusters_new.labels_
```

```python
# Assign clusters to data set

data["n_clustersid"]=clusters_new.labels_
clusters_new.cluster_centers_
```

```python
data.groupby("n_clustersid").agg(["mean"]).reset_index()
```

# DBSCAN Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("G:\my Assignemt\Assignment 7 Clustering\crime_data.csv")
data
```

```python
data.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
array = data.values
array
```

```python
# Normalizing
scalar = StandardScaler().fit(array)
scaled_data_df = scaler.transform(array)
scaled_data_df
```

```python
from sklearn.neighbors import NearestNeighbors
import seaborn as sns
sns.set()

neighbors = NearestNeighbors(n_neighbors=6)
fitneighbors = neighbors.fit(scaled_data_df)
distances, indices = fitneighbors.kneighbors(scaled_data_df)
```

```python
distances = np.sort(distances,axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
dbscan = DBSCAN(eps=1.2,min_samples=4)
dbscan.fit(scaled_data_df)
```

```python
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_,columns=["Clusters"])
cl
```

```python
import sklearn
sklearn.metrics.silhouette_score(scaled_data_df,dbscan.labels_)
```

```python
pd.concat([data,cl],axis=1)
```
